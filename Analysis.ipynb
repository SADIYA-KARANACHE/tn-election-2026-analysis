```python
df2021_sorted = df2021.sort_values(
    ['ac_number', 'votes'],
    ascending=[True, False]
)
```

```python
df2021_sorted.head(20)
```

```python
winners2021 = df2021_sorted.groupby('ac_number').first().reset_index()
```

```python
winners2021.head()
```

```python
winners2021.shape
```

```python
runnerups2021 = df2021_sorted.groupby('ac_number').nth(1).reset_index()
```

```python
runnerups2021.head()
```

```python
runnerups2021.shape
```

```python
winners2021 = winners2021.rename(columns={
    'candidate': 'winner_candidate',
    'party': 'winner_party',
    'votes': 'winner_votes'
})
```

```python
runnerups2021 = runnerups2021.rename(columns={
    'candidate': 'runnerup_candidate',
    'party': 'runnerup_party',
    'votes': 'runnerup_votes'
})
```

```python
final2021 = pd.merge(
    winners2021[
        ['ac_number', 'constituency', 'winner_candidate', 'winner_party', 'winner_votes']
    ],
    
    runnerups2021[
        ['ac_number', 'runnerup_candidate', 'runnerup_party', 'runnerup_votes']
    ],
    
    on='ac_number'
)
```

```python
final2021.head()
```

```python
final2021['margin'] = (
    final2021['winner_votes'] -
    final2021['runnerup_votes']
)
```

```python
final2021.head()
```

```python
final2021.shape
```

```python
compare2021 = final2021[
    ['ac_number', 'constituency', 'winner_party']
].rename(columns={
    'winner_party': 'winner_party_2021'
})
```

```python
compare2026 = final2026[
    ['ac_number', 'winner_party']
].rename(columns={
    'winner_party': 'winner_party_2026'
})
```

```python
flip_analysis = pd.merge(
    compare2021,
    compare2026,
    on='ac_number'
)
```

```python
flip_analysis['flipped'] = (
    flip_analysis['winner_party_2021'] !=
    flip_analysis['winner_party_2026']
)
```

```python
flip_analysis.head()
```

```python
flip_analysis['flipped'].sum()
```

```python
flipped_seats = flip_analysis[
    flip_analysis['flipped'] == True
]
```

```python
final2021 = pd.merge(
    final2021,
    
    master[['ac_number', 'region', 'reserved']],
    
    on='ac_number',
    how='left'
)
```

```python
final2026 = pd.merge(
    final2026,
    
    master[['ac_number', 'region', 'reserved']],
    
    on='ac_number',
    how='left'
)
```

```python
final2021.head()
```

```python
final2026.head()
```

```python
region_2021 = (
    final2021
    .groupby(['region', 'winner_party'])
    .size()
    .reset_index(name='seats_won')
)
```

```python
region_2026 = (
    final2026
    .groupby(['region', 'winner_party'])
    .size()
    .reset_index(name='seats_won')
)
```

```python
region_2026.head()
```

```python
region_2021_sorted = region_2021.sort_values(
    ['region', 'seats_won'],
    ascending=[True, False]
)
```

```python
region_2026_sorted = region_2026.sort_values(
    ['region', 'seats_won'],
    ascending=[True, False]
)
```

```python
region_2026_sorted
```

```python
top_region_2021 = (
    region_2021_sorted
    .groupby('region')
    .first()
    .reset_index()
)
```

```python
top_region_2026 = (
    region_2026_sorted
    .groupby('region')
    .first()
    .reset_index()
)
```
